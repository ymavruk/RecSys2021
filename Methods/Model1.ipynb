# Hybrid Model using game tags

First only content based filtering

```python
#get the data 
csvPath = "/Users/zein/Desktop/dataWithRatings.csv"
```

```python
import pandas as pd
df = pd.read_csv (csvPath)
```

### Select a random user

```python
#select a random user that has played more that minHours #JUST LIKE LAB1
minHours = 150
users_ratings = df.groupby(['user_id']).count()
selected = users_ratings['hours'] > minHours
selected_users = users_ratings.loc[selected]
random_selected = selected_users.sample() 
select_column_df = random_selected.reset_index()['user_id'] 
selected_user = select_column_df.iloc[0] 
print("Selected user: " + str(selected_user))
```

```python
selected_user_ratings = df.loc[df['user_id'] == selected_user]
# selected_user_ratings = selected_user_ratings.sort_values(by='hours', ascending=True)
print("Rated games: " + str(selected_user_ratings.shape[0]))
display(selected_user_ratings.head(10))
```

### Rated games by the user

```python
rated_games_df = selected_user_ratings
rated_games_df = rated_games_df[['name', 'popular_tags']]
print("Rated games: " + str(rated_games_df.shape[0]))
display(rated_games_df.head(10))
```

### Unrated games by the user

```python
diff = set(df.index) - set(rated_games_df.index)
unrated_games_df = df.loc[diff]
unrated_games_df = unrated_games_df[['name', 'popular_tags']]
unrated_games_df = unrated_games_df.drop_duplicates(subset ="name") #Had to remove duplicates 
print("Unrated games: " + str(unrated_games_df.shape[0]))
display(unrated_games_df.head(10))
```

### Combine the rated games to their ratings

```python
rated_games_df = rated_games_df.join(selected_user_ratings.set_index('name')['Rating M1'], on='name')
rated_games_df = rated_games_df.join(selected_user_ratings.set_index('name')['Rating M2'], on='name')
print("Rated games: " + str(rated_games_df.shape[0]))
display(rated_games_df.head(10))
```

## Predict the ratings of the unrated games using KNN

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(rated_games_df['popular_tags'])
print(X.shape)
```

```python
from sklearn.neighbors import KNeighborsRegressor

y = rated_games_df['Rating M1']
neigh = KNeighborsRegressor(n_neighbors=5)
neigh.fit(X, y)
```

```python
X_unrated = vectorizer.transform(unrated_games_df['popular_tags'].values.astype('U'))
print(X_unrated.shape)
y_unrated = neigh.predict(X_unrated)
print(y_unrated.shape)
```

```python
unrated_games_df['predicted_ratings_KNN M1'] = y_unrated
unrated_games_df_KNN_ordering = unrated_games_df.sort_values(by='predicted_ratings_KNN M1', ascending=False)
```

```python
y = rated_games_df['Rating M2']
neigh = KNeighborsRegressor(n_neighbors=5)
neigh.fit(X, y)
```

```python
X_unrated = vectorizer.transform(unrated_games_df['popular_tags'].values.astype('U'))
print(X_unrated.shape)
y_unrated = neigh.predict(X_unrated)
print(y_unrated.shape)
```

### Predicted Ratings

```python
unrated_games_df['predicted_ratings_KNN M2'] = y_unrated
unrated_games_df_KNN_ordering = unrated_games_df.sort_values(by='predicted_ratings_KNN M2', ascending=False)
display(unrated_games_df_KNN_ordering.head(10))
```
